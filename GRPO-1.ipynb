```python
# RobustModelParser.py
import yaml
import re
from pathlib import Path

class RobustModelParser:
    """
    Extracts variables, shocks, parameters, calibration, equations,
    and labels from YAML and comment-style lines.
    """

    def __init__(self, file_path):
        self.file_path = Path(file_path)
        self.raw_text = self.file_path.read_text()
        self.yaml_obj = yaml.safe_load(self.raw_text)

        self.comment_labels = self._extract_comment_labels()
        self.yaml_labels = self.yaml_obj.get("labels", {}) or {}
        self.labels = {**self.comment_labels, **self.yaml_labels}

    def _extract_comment_labels(self):
        pattern = r"#\s*(.*?)\s{2,}([A-Za-z0-9_]+)\s*$"
        matches = re.findall(pattern, self.raw_text, flags=re.MULTILINE)
        return {var.strip(): desc.strip() for desc, var in matches}

    def symbols(self):
        return self.yaml_obj.get("symbols", {}) or {}

    def variables(self):
        return self.symbols().get("variables", []) or []

    def measurement_variables(self):
        return self.symbols().get("measurement_variables", []) or []

    def shocks(self):
        return self.symbols().get("shocks", []) or []

    def measurement_shocks(self):
        return self.symbols().get("measurement_shocks", []) or []

    def parameters(self):
        return self.symbols().get("parameters", []) or []

    def calibration(self):
        return self.yaml_obj.get("calibration", {}) or {}

    def equations(self):
        return (
            self.yaml_obj.get("equations", []) or [],
            self.yaml_obj.get("measurement_equations", []) or []
        )

    def build_llm_context(self):
        eqs, meqs = self.equations()
        ctx = f"MODEL CONTEXT — {self.file_path.name}\n\n"

        def add(title, arr):
            nonlocal ctx
            ctx += f"--- {title} ---\n"
            for x in arr:
                label = self.labels.get(x, "(no label)")
                ctx += f"  - {x}: {label}\n"
            ctx += "\n"

        add("VARIABLES", self.variables())
        add("MEASUREMENT VARIABLES", self.measurement_variables())
        add("PARAMETERS", self.parameters())
        add("SHOCKS", self.shocks())
        add("MEASUREMENT SHOCKS", self.measurement_shocks())

        ctx += "--- CALIBRATION ---\n"
        for k, v in self.calibration().items():
            ctx += f"  - {k}: {v}\n"

        ctx += "\n--- TRANSITION EQUATIONS ---\n"
        for e in eqs:
            ctx += f"  - {e}\n"

        ctx += "\n--- MEASUREMENT EQUATIONS ---\n"
        for e in meqs:
            ctx += f"  - {e}\n"

        return ctx
```

```python
# generic_dsge_simulator.py
import copy
import io
import sys
from typing import Optional, List, Union, Dict, Any

import numpy as np
import pandas as pd

# Snowdrop imports (assumes snowdrop is on PYTHONPATH)
from snowdrop.src.driver import importModel, run
from snowdrop.src.model.util import setValues

# ---------------------------
# Generic DSGE Simulator (final)
# ---------------------------
class GenericDSGESimulator:
    """
    Robust, model-agnostic wrapper around Snowdrop models.

    Features:
      - inject_shocks supports: scalar, list, list-of-(t,val), pandas.Series
      - update_parameters supports scalar and time-varying calibrated arrays (via setValues fallback)
      - reset_shocks to zero the shock matrix for current horizon
      - clone() to create an identical simulator preserving calibration/options (deep copy)
      - auto solver selection based on model.options["linear"]
      - generate_irf() to compute IRF = shocked - baseline using only the public API
      - rl_step() safe single-step interface for RL
    """

    def __init__(self, model_path: str, periods: int = 40, silence_snowdrop: bool = True):
        self.model_path = str(model_path)
        self.periods = int(periods)

        # optionally suppress Snowdrop import-time prints
        _stdout = sys.stdout
        if silence_snowdrop:
            sys.stdout = io.StringIO()
        try:
            self.model = importModel(self.model_path)
        finally:
            if silence_snowdrop:
                sys.stdout = _stdout

        # ensure T/periods is set (Snowdrop uses options["T"])
        self.model.options["T"] = self.periods
        self.model.options.setdefault("range", None)

        # cache symbol lists
        s = getattr(self.model, "symbols", {})
        self.vars = list(s.get("variables", []))
        self.shock_names = list(s.get("shocks", []))
        self.param_names = list(s.get("parameters", []))

    # ---------------------------
    # Helper: choose default solver based on model linearity
    # ---------------------------
    def _choose_solver(self, solver: Optional[str]) -> str:
        if solver is not None:
            return solver

        # Snowdrop convention: model.options['linear'] frequently exists.
        is_linear = self.model.options.get("linear", True)
        if is_linear:
            # Linear solver default
            return "BinderPesaran"
        else:
            return "LBJ"

    # ---------------------------
    # Shock injection (robust)
    # ---------------------------
    def inject_shocks(self, shock_dict: Dict[str, Any], reset: bool = True):
        """
        Inject shocks into the model.

        shock_dict formats accepted:
          - scalar: {"epsr": 0.1}
          - list:   {"epsr": [0.1, 0.0, ...]}
          - tuples: {"em": [(0,0.02),(3,-0.01)]}
          - pandas.Series indexed by 0..T-1 (or with datetime index, but we use positional)
        """
        import pandas as _pd

        T = int(self.model.options.get("T", self.periods))
        nshk = len(self.shock_names)
        S = np.zeros((T, nshk))

        for shk, values in shock_dict.items():
            if shk not in self.shock_names:
                print(f"⚠️ Warning: unknown shock '{shk}' — ignoring.")
                continue
            j = self.shock_names.index(shk)

            # scalar
            if np.isscalar(values):
                S[0, j] = float(values)
                continue

            # pandas Series
            if isinstance(values, _pd.Series):
                for idx, val in enumerate(values.iloc[:T]):
                    S[idx, j] = float(val)
                continue

            # list of tuples (t, val)
            if isinstance(values, list) and len(values) > 0 and isinstance(values[0], tuple):
                for (t, val) in values:
                    ti = int(t)
                    if 0 <= ti < T:
                        S[ti, j] = float(val)
                continue

            # list of values (path)
            if isinstance(values, list):
                # truncate/pad
                for i, val in enumerate(values[:T]):
                    S[i, j] = float(val)
                continue

            raise ValueError(f"Invalid shock format for {shk}: {values}")

        # write back to model options and calibration if present
        self.model.options["shock_values"] = S
        try:
            # some models expect calibration["shocks"]
            self.model.calibration["shocks"] = S
        except Exception:
            pass

    # ---------------------------
    # Reset shocks to zero for the current T horizon
    # ---------------------------
    def reset_shocks(self):
        T = int(self.model.options.get("T", self.periods))
        nshk = len(self.shock_names)
        zeros = np.zeros((T, nshk))
        self.model.options["shock_values"] = zeros
        try:
            self.model.calibration["shocks"] = zeros
        except Exception:
            pass

    # ---------------------------
    # Update parameters (robust)
    # ---------------------------
    def update_parameters(self, param_dict: Dict[str, Any], start: Optional[Any] = None):
        """
        Uses setValues for robust mapping when available, otherwise updates direct calibration.
        """
        if not param_dict:
            return

        # Try using Snowdrop's setValues to handle arrays/time-varying parameters
        try:
            curr = self.model.calibration.get("parameters", None)
            new_calib, _, _ = setValues(
                model=self.model,
                d=param_dict,
                names=self.param_names,
                values=curr,
                start=start,
                isShock=False
            )
            self.model.calibration["parameters"] = new_calib
            return
        except Exception:
            # fallback: direct overwrite where possible
            param_values = self.model.calibration.get("parameters", None)
            if param_values is None:
                raise RuntimeError("Model calibration['parameters'] not found and setValues failed.")
            for name, new_val in param_dict.items():
                if name in self.param_names:
                    idx = self.param_names.index(name)
                    val = param_values[idx]
                    if np.isscalar(val):
                        param_values[idx] = new_val
                    else:
                        # assume array-like
                        param_values[idx][:] = new_val
            self.model.calibration["parameters"] = param_values

    # ---------------------------
    # Run simulation (suppresses Snowdrop prints and returns DataFrame)
    # ---------------------------
    def run(self, solver: Optional[str] = None, output_vars: Optional[List[str]] = None, silent_print: bool = False) -> pd.DataFrame:
        """
        Run the model and return a DataFrame with result variables (or subset output_vars).
        This method suppresses Snowdrop's internal print output.
        """
        solver_to_use = self._choose_solver(solver)

        # suppress prints
        _stdout = sys.stdout
        sys.stdout = io.StringIO()
        try:
            # Some Snowdrop versions accept "output_variables" kwarg, but to be robust we'll call run normally
            y, rng = run(
                model=self.model,
                Solver=solver_to_use,
                Plot=False,
                Output=False
            )
        finally:
            sys.stdout = _stdout

        # ensure y is an ndarray and rng is indexable
        rows = min(len(rng), getattr(y, "shape", [0])[0])
        df = pd.DataFrame(y[:rows], index=rng[:rows], columns=self.vars)

        # if user requested subset of vars, validate and return subset
        if output_vars:
            missing = [v for v in output_vars if v not in df.columns]
            if missing:
                raise ValueError(f"Requested output_vars not present in model: {missing}")
            df_print = df[output_vars]
        else:
            df_print = df

        # print only when not silent_print
        if not silent_print:
            print("\n=== SIMULATION OUTPUT (subset) ===")
            print(df_print.head(12))
            print("\n=== STATISTICS (subset) ===")
            print(df_print.describe())

        return df_print

    # ---------------------------
    # Clone current simulator (deep copy of model state) — returns new GenericDSGESimulator
    # ---------------------------
    def clone(self):
        """
        Create a fresh simulator instance with the SAME model_path and periods,
        then deep-copy calibration and options so the clone has the same current state.
        """
        new_sim = GenericDSGESimulator(self.model_path, periods=self.periods, silence_snowdrop=True)
        # deep copy current model's calibration and options into the new model
        new_sim.model.calibration = copy.deepcopy(getattr(self.model, "calibration", {}))
        new_sim.model.options = copy.deepcopy(getattr(self.model, "options", {}))

        # ensure new_sim caches are updated (symbols same)
        s = getattr(new_sim.model, "symbols", {})
        new_sim.vars = list(s.get("variables", []))
        new_sim.shock_names = list(s.get("shocks", []))
        new_sim.param_names = list(s.get("parameters", []))

        return new_sim

    # ---------------------------
    # IRF generator (model-agnostic)
    # ---------------------------
    def generate_irf(
        self,
        shock: str,
        size: float = 1.0,
        horizon: Optional[int] = None,
        vars: Optional[List[str]] = None,
        shock_at: int = 0,
        solver: Optional[str] = None,
        use_shock_std: bool = False,
        silent_print: bool = False
    ) -> pd.DataFrame:
        """
        Compute IRF = (shocked simulation) - (baseline simulation).

        Args:
          shock: name of shock in model.symbols['shocks']
          size: magnitude of shock
          horizon: #periods for IRF (defaults to self.periods)
          vars: list of variable names to return (subset). If None returns all.
          shock_at: integer period index where shock is applied (default t=0)
          solver: optional override solver
          use_shock_std: if True, look up calibration key 'std_<shock>' and multiply size by it

        Returns:
          DataFrame indexed by simulation dates, columns = requested variables; values = shocked - baseline
        """
        if horizon is None:
            horizon = self.periods
        horizon = int(horizon)

        if shock not in self.shock_names:
            raise ValueError(f"Shock '{shock}' not in model.shocks: {self.shock_names}")

        # optionally scale by calibrated std
        mag = float(size)
        if use_shock_std:
            std_key = f"std_{shock}"
            std_val = None
            calib = getattr(self.model, "calibration", {})
            if isinstance(calib, dict) and std_key in calib:
                std_val = calib[std_key]
            if std_val is None:
                print(f"⚠️ WARNING: std_{shock} not found in calibration; proceeding with raw size.")
            else:
                mag = mag * float(std_val)

        # --------------------------
        # Run baseline (clone and reset shocks)
        # --------------------------
        base_sim = self.clone()
        base_sim.model.options["T"] = horizon
        base_sim.reset_shocks()
        baseline_df = base_sim.run(solver=solver, output_vars=vars, silent_print=silent_print)

        # --------------------------
        # Run shocked sim (clone current state so param changes are preserved)
        # --------------------------
        shocked_sim = self.clone()
        shocked_sim.model.options["T"] = horizon
        shocked_sim.reset_shocks()
        shocked_sim.inject_shocks({shock: [(shock_at, mag)]})
        shocked_df = shocked_sim.run(solver=solver, output_vars=vars, silent_print=silent_print)

        # align indices/columns
        baseline_df, shocked_df = baseline_df.align(shocked_df, join="outer", axis=0)
        irf_df = shocked_df.fillna(0) - baseline_df.fillna(0)
        irf_df = irf_df.iloc[:horizon]

        if not silent_print:
            print(f"\n=== IRF for shock '{shock}' (size={mag}) ===")
            print(irf_df.head(12))

        return irf_df

    # ---------------------------
    # RL STEP (safe version)
    # ---------------------------
    def rl_step(self, action_shocks: Dict[str, Any], param_updates: Optional[Dict[str, Any]] = None, output_vars: Optional[List[str]] = None, solver: Optional[str] = None):
        """
        Performs a single-step transition for RL agents.

        This:
            - clones the simulator (so original T is preserved)
            - runs a T=1 simulation with the given shock and parameters
            - returns the single-step resulting state (last row)
        """
        temp = self.clone()
        temp.model.options["T"] = 1

        # apply RL-driven parameter updates (policy changes) to the temp sim
        if param_updates:
            temp.update_parameters(param_updates)

        # place action shocks at t=0
        temp.reset_shocks()
        temp.inject_shocks(action_shocks)

        # run one-step model and return last row
        df = temp.run(solver=solver, output_vars=output_vars, silent_print=True)
        # df might be empty or single-row; return a Series or empty frame accordingly
        if df.shape[0] == 0:
            return df
        return df.iloc[-1]
```

```python
# ===================== CELL: COMPREHENSIVE DIAGNOSTIC SUITE =====================
import torch
import numpy as np
import json
import re
import copy
import yaml
import os
import traceback
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

# --- 1. YAML Patch (Prevent RBC Crash) ---
def dummy_constructor(loader, node): return None
yaml.SafeLoader.add_constructor("!MvNormal", dummy_constructor)


# --- 2. ADVANCED CONFIGURATION: Models, Crises, & Expanded Actions ---

# Define the "World" of simulators
GAME_CONFIGS = [
    # 1. QPM: The Standard Central Bank Model
    {
        "name": "QPM_Full_Control",
        "model": "/content/model.yaml",
        "output_vars": ["L_GDP_GAP", "DLA_CPI", "RS"],
        "scenarios": [
            ("Stagflation (Supply Shock)", {"SHK_DLA_CPI": [2.5]*8, "SHK_L_GDP_GAP": [-1.0]*8}),
            ("Demand Collapse (Recession)", {"SHK_L_GDP_GAP": [-3.0]*8})
        ],
        # EXPANDED ACTIONS: Rates, Real Rate Trend, Inflation Target
        "allowed_shocks": ["SHK_RS", "SHK_RR_BAR", "SHK_D4L_CPI_TAR"],
        "constraints": {"SHK_RS": (-3, 3), "SHK_RR_BAR": (-2, 2), "SHK_D4L_CPI_TAR": (-1, 1)}
    },

    # 2. SW: Smets-Wouters (Business Cycles)
    {
        "name": "SW_Business_Cycles",
        "model": "/content/sw_model.yaml",
        "output_vars": ["y", "pinf", "r"],
        "scenarios": [
            ("Productivity Crash", {"ea": [-2.0]*8}), # TFP Shock
            ("Cost Push Inflation", {"epinf": [3.0]*8}) # Price Markup Shock
        ],
        # EXPANDED ACTIONS: Monetary (em), Fiscal (eg - Spending), Structural (ea - Tech investment)
        "allowed_shocks": ["em", "eg", "ea"],
        "constraints": {"em": (-2.5, 2.5), "eg": (-2.0, 2.0), "ea": (0, 1.0)}
    },

    # 3. GSW: Gali-Smets-Wouters (COVID/Pandemic Focus)
    {
        "name": "GSW_Pandemic_Sim",
        "model": "/content/gsw_model.yaml",
        "output_vars": ["y", "unempl", "pinf"],
        "scenarios": [
            # The "Lockdown": Labor Supply (els) and Productivity (ea) collapse
            ("COVID Lockdown", {"els": [-5.0]*8, "ea": [-3.0]*8})
        ],
        # EXPANDED ACTIONS: Monetary (em), Wage Support (ew), Gov Spending (eg)
        "allowed_shocks": ["em", "eg", "ew"],
        "constraints": {"em": (-2, 2), "eg": (0, 5.0), "ew": (-2, 2)}
    },

    # 4. Ireland 2004: New Keynesian Tech Shocks
    {
        "name": "Ireland_NK_Tech",
        "model": "/content/Ireland2004.yaml",
        "output_vars": ["y", "pie", "r"],
        "scenarios": [
            ("Tech Collapse", {"epsa": [-3.0]*8}),
            ("Preference Shock", {"epse": [2.0]*8})
        ],
        # ACTIONS: Rates (epsr), Tech Investment (epsz)
        "allowed_shocks": ["epsr", "epsz"],
        "constraints": {"epsr": (-2, 2), "epsz": (-1, 1)}
    },

    # 5. RBC: Real Business Cycle (Growth Focus)
    {
        "name": "RBC_Growth",
        "model": "/content/RBC.yaml",
        "output_vars": ["Y", "C", "r"],
        "scenarios": [
            ("Capital Destruction", {"ea": [-4.0]*8}) # Negative TFP
        ],
        # ACTIONS: RBC has no central bank. We simulate "Structural Reform" via 'ea' (TFP boost)
        "allowed_shocks": ["ea"],
        "constraints": {"ea": (0, 2.0)} # Can only IMPROVE tech
    },

    # 6. MVF: Multivariate Filter (Gap Estimation)
    {
        "name": "MVF_US_Gap",
        "model": "/content/MVF_US.yaml",
        "output_vars": ["Y", "PIE", "UNR_GAP"],
        "scenarios": [
            ("Overheating", {"RES_Y": [3.0]*8, "RES_PIE": [1.5]*8})
        ],
        # ACTIONS: Demand Management (RES_Y)
        "allowed_shocks": ["RES_Y"],
        "constraints": {"RES_Y": (-3, 3)}
    }
]

# --- 3. HELPER FUNCTIONS ---

def parse_action_robust(text: str, allowed_keys: list) -> dict:
    """Extracts JSON or Key-Value pairs robustly."""
    text = text.strip()
    extracted = {}

    # Method A: Regex for "Key": Value or Key: Value
    for key in allowed_keys:
        # Matches: "key": 1.2, key: -0.5, key = 5
        pattern = r'["\']?' + re.escape(key) + r'["\']?\s*[:=]\s*([-+]?\d*\.?\d+)'
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            extracted[key] = float(match.group(1))

    # Method B: JSON Search if Regex failed to find everything
    if len(extracted) < 1:
        try:
            match = re.search(r"\{.*\}", text, re.DOTALL)
            if match:
                clean_json = re.sub(r",\s*}", "}", match.group(0)) # Fix trailing commas
                data = json.loads(clean_json)
                # Flatten
                if isinstance(data, dict):
                    for k, v in data.items():
                        if k in allowed_keys: extracted[k] = float(v)
                        elif isinstance(v, dict):
                            for nk, nv in v.items():
                                if nk in allowed_keys: extracted[nk] = float(nv)
        except:
            pass
    return extracted

def merge_shocks_history(crisis_shocks, agent_action, burn_in_len, shock_names):
    """Combines historical crisis with new agent action."""
    final_dict = {}
    # 1. Crisis History
    for k, v_list in crisis_shocks.items():
        final_dict[k] = [(t, val) for t, val in enumerate(v_list)]

    # 2. Agent Action (Applied at t=burn_in_len)
    for k, val in agent_action.items():
        if k in shock_names:
            if k not in final_dict: final_dict[k] = []
            final_dict[k].append((burn_in_len, float(val)))
            # OPTIONAL: Persistence? Let's assume action lasts 4 quarters for impact
            for p in range(1, 4):
                 final_dict[k].append((burn_in_len + p, float(val) * 0.8)) # Decay

    return final_dict

def generate_prompt(game, scenario_name, state, parser):
    # Dynamic Example based on allowed shocks
    ex_key = game["allowed_shocks"][0]
    ex_json = json.dumps({ex_key: 0.5})

    state_str = ", ".join([f"{k}: {float(state.get(k,0)):.2f}" for k in game["output_vars"]])
    actions_str = ", ".join(game["allowed_shocks"])

    prompt = (
        f"<|im_start|>system\n"
        f"You are an Economic Policy AI. The economy is in a '{scenario_name}'.\n"
        f"Goal: Stabilize output variables to 0.0.\n"
        f"Output: A JSON object containing Policy Adjustments.\n"
        f"Example Format: {ex_json}\n"
        f"<|im_end|>\n"
        f"<|im_start|>user\n"
        f"Model: {game['name']}\n"
        f"Current Crisis State: {state_str}\n"
        f"Available Policy Levers: {actions_str}\n"
        f"Constraints: {json.dumps(game['constraints'])}\n\n"
        f"Task: Choose policy values to reverse the crisis.\n"
        f"Action JSON:\n<|im_end|>\n<|im_start|>assistant\n"
    )
    return prompt

# --- 4. DIAGNOSTIC RUNNER ---

def run_comprehensive_diagnostic(model, tokenizer, group_size=4):
    print(f"\n{'='*80}")
    print(f"🏥 COMPREHENSIVE DIAGNOSTIC: {len(GAME_CONFIGS)} Models")
    print(f"{'='*80}")

    for config in GAME_CONFIGS:
        print(f"\n👇 MODEL: {config['name']}")

        # 1. Load Sim
        try:
            # Auto-fix path
            path = config["model"]
            if not os.path.exists(path) and os.path.exists(f"/content/{path}"):
                path = f"/content/{path}"

            sim = GenericDSGESimulator(path, periods=20, silence_snowdrop=True)
            parser = RobustModelParser(path)
        except Exception as e:
            print(f"   ❌ Skipped (Load Error): {e}")
            continue

        # 2. Iterate Scenarios
        for s_name, crisis_shocks in config["scenarios"]:
            print(f"   🚩 SCENARIO: {s_name}")

            # Burn-in
            burn_in = 8
            sim.reset_shocks()
            sim.inject_shocks(crisis_shocks)
            sim.model.options["T"] = burn_in
            try:
                df_burn = sim.run(output_vars=config["output_vars"], silent_print=True)
            except:
                print("      ❌ Burn-in failed.")
                continue

            current_state = df_burn.iloc[-1]
            print(f"      State (t={burn_in}): {current_state.to_dict()}")

            # Generate
            prompt = generate_prompt(config, s_name, current_state, parser)
            inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

            with torch.no_grad():
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=45,
                    do_sample=True,
                    temperature=1.2, # Force diversity
                    top_p=0.9,
                    num_return_sequences=group_size,
                    pad_token_id=tokenizer.eos_token_id
                )

            decoded = tokenizer.batch_decode(outputs[:, inputs.input_ids.shape[1]:], skip_special_tokens=True)

            # Verify Diversity
            actions_taken = []
            rewards = []

            print(f"      🏃 Simulating {group_size} trajectories...")

            for i, text in enumerate(decoded):
                # Parse
                raw_act = parse_action_robust(text, config["allowed_shocks"])

                # Clone & Run
                sim_clone = sim.clone()
                sim_clone.model.options["T"] = burn_in + 12
                merged = merge_shocks_history(crisis_shocks, raw_act, burn_in, sim.shock_names)
                sim_clone.reset_shocks()
                sim_clone.inject_shocks(merged)

                df_traj = sim_clone.run(output_vars=config["output_vars"], silent_print=True)

                # Score
                loss = 0
                for v in config["output_vars"]:
                    vals = df_traj[v].iloc[burn_in:].values
                    loss += np.sum(vals**2)

                actions_taken.append(raw_act)
                rewards.append(-loss)

                # Sample Print
                if i == 0:
                    print(f"         Sample Prompt: {prompt[-150:].replace(chr(10), ' ')}")
                    print(f"         Sample Raw Out: {text.strip().replace(chr(10), ' ')}")
                    print(f"         Sample Parsed: {raw_act}")

            # STATISTICS
            rewards = np.array(rewards)
            std_r = np.std(rewards)

            if std_r > 1e-5:
                best_idx = np.argmax(rewards)
                print(f"      ✅ SUCCESS: Variance Detected (Std: {std_r:.2f}).")
                print(f"         Best Action: {actions_taken[best_idx]} (Reward: {rewards[best_idx]:.2f})")
                print(f"         Worst Action: {actions_taken[np.argmin(rewards)]} (Reward: {rewards.min():.2f})")
            else:
                print(f"      ⚠️ WARNING: No Variance (Rewards identical: {rewards[0]:.2f}).")
                if len(actions_taken) > 0 and actions_taken[0] == {}:
                    print("         -> Issue: Parser found NO actions.")
                else:
                    print("         -> Issue: Simulator physics unresponsive to these shocks.")

# --- EXECUTE ---
run_comprehensive_diagnostic(model, tokenizer)
```

```python
# ===================== CELL: ROBUST GRPO TRAINING (With Inaction Penalty) =====================
import torch
import torch.nn.functional as F
import numpy as np
import json
import re
import copy
import yaml
import os
import traceback
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import LoraConfig, get_peft_model, TaskType

# --- 1. CONFIGURATION ---
DRIVE_PATH = "/content/drive/MyDrive/Economic_Agent_RL"
os.makedirs(DRIVE_PATH, exist_ok=True)

MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"
GRPO_GROUP_SIZE = 4
EPOCHS = 3
LEARNING_RATE = 1e-5
INACTION_PENALTY = -5000.0  # <--- NEW: Force agent to act

# Fix YAML
def dummy_constructor(loader, node): return None
yaml.SafeLoader.add_constructor("!MvNormal", dummy_constructor)


# --- 2. GAME CONFIGS (Same as before) ---
GAME_CONFIGS = [
    {
        "name": "QPM_Full_Control",
        "model": "model.yaml",
        "output_vars": ["L_GDP_GAP", "DLA_CPI", "RS"],
        "scenarios": [
            ("Stagflation", {"SHK_DLA_CPI": [2.5]*8, "SHK_L_GDP_GAP": [-1.0]*8}),
            ("Recession", {"SHK_L_GDP_GAP": [-3.0]*8})
        ],
        "allowed_shocks": ["SHK_RS", "SHK_RR_BAR", "SHK_D4L_CPI_TAR"],
        "constraints": {"SHK_RS": (-3, 3), "SHK_RR_BAR": (-2, 2), "SHK_D4L_CPI_TAR": (-1, 1)}
    },
    {
        "name": "SW_Business_Cycles",
        "model": "sw_model.yaml",
        "output_vars": ["y", "pinf", "r"],
        "scenarios": [("Productivity Crash", {"ea": [-5.0]*8}), ("Cost Push", {"epinf": [3.0]*8})],
        "allowed_shocks": ["em", "eg", "ea"],
        "constraints": {"em": (-3.0, 3.0), "eg": (-3.0, 3.0), "ea": (0, 2.0)}
    },
    {
        "name": "GSW_Pandemic_Sim",
        "model": "gsw_model.yaml",
        "output_vars": ["y", "unempl", "pinf"],
        "scenarios": [("COVID Lockdown", {"els": [10.0]*8})],
        "allowed_shocks": ["em", "eg", "ew"],
        "constraints": {"em": (-2, 2), "eg": (0, 8.0), "ew": (-2, 2)}
    },
    {
        "name": "Ireland_NK_Tech",
        "model": "Ireland2004.yaml",
        "output_vars": ["y", "pie", "r"],
        "scenarios": [("Tech Collapse", {"epsa": [-3.0]*8})],
        "allowed_shocks": ["epsr", "epsz"],
        "constraints": {"epsr": (-2, 2), "epsz": (-1, 1)}
    },
    {
        "name": "MVF_US_Gap",
        "model": "MVF_US.yaml",
        "output_vars": ["Y", "PIE", "UNR_GAP"],
        "scenarios": [("Overheating", {"RES_Y": [3.0]*8, "RES_PIE": [1.5]*8})],
        "allowed_shocks": ["RES_Y"],
        "constraints": {"RES_Y": (-3, 3)}
    }
]

# --- 3. HELPER FUNCTIONS ---
def parse_action_robust(text: str, allowed_keys: list) -> dict:
    text = text.strip()
    extracted = {}

    # 1. Regex for Key-Value
    for key in allowed_keys:
        pattern = r'["\']?' + re.escape(key) + r'["\']?\s*[:=]\s*([-+]?\d*\.?\d+)'
        match = re.search(pattern, text, re.IGNORECASE)
        if match: extracted[key] = float(match.group(1))

    # 2. JSON Fallback
    if not extracted:
        try:
            match = re.search(r"\{.*\}", text, re.DOTALL)
            if match:
                clean_json = re.sub(r",\s*}", "}", match.group(0))
                data = json.loads(clean_json)
                if isinstance(data, dict):
                    for k, v in data.items():
                        if k in allowed_keys: extracted[k] = float(v)
        except: pass
    return extracted

def constrain_action(action, allowed, constraints):
    out = {}
    for k, v in action.items():
        if k in allowed:
            try:
                val = float(v)
                mn, mx = constraints.get(k, (-2.5, 2.5))
                out[k] = max(mn, min(mx, val))
            except: pass
    return out

def merge_shocks_history(crisis_shocks, agent_action, burn_in_len, shock_names):
    final_dict = {}
    for k, v_list in crisis_shocks.items():
        final_dict[k] = [(t, val) for t, val in enumerate(v_list)]
    for k, val in agent_action.items():
        if k in shock_names:
            if k not in final_dict: final_dict[k] = []
            final_dict[k].append((burn_in_len, float(val)))
            for p in range(1, 4):
                 final_dict[k].append((burn_in_len + p, float(val) * 0.8))
    return final_dict

# --- 4. MODEL LOADING ---
# Check if checkpoint exists to resume
last_ckpt = f"{DRIVE_PATH}/checkpoint_epoch_2"
if os.path.exists(last_ckpt):
    print(f"🔄 Resuming from Checkpoint: {last_ckpt}")
    tokenizer = AutoTokenizer.from_pretrained(last_ckpt, trust_remote_code=True)
    base_model = AutoModelForCausalLM.from_pretrained(MODEL_ID, torch_dtype=torch.float16, device_map="auto", trust_remote_code=True)
    model = PeftModel.from_pretrained(base_model, last_ckpt, is_trainable=True)
else:
    print("⚙️ Initializing New Model...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, trust_remote_code=True)
    tokenizer.pad_token = tokenizer.eos_token
    base_model = AutoModelForCausalLM.from_pretrained(MODEL_ID, torch_dtype=torch.float16, device_map="auto", trust_remote_code=True)
    peft_config = LoraConfig(task_type=TaskType.CAUSAL_LM, r=16, lora_alpha=32, lora_dropout=0.05, target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"])
    model = get_peft_model(base_model, peft_config)

optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)
print("✅ Model Ready.")

# --- 5. ROBUST TRAINING LOOP ---

def train_one_step(game, scenario_name, crisis_shocks, epoch_idx):
    # A. Setup
    try:
        path = game["model"]
        if not os.path.exists(path) and os.path.exists(f"/content/{path}"): path = f"/content/{path}"
        sim = GenericDSGESimulator(path, periods=25, silence_snowdrop=True)
        if "shock_values" in sim.model.options: del sim.model.options["shock_values"]
        parser = RobustModelParser(path)
    except: return

    # B. Burn-in
    burn_in_len = 8
    sim.reset_shocks()
    sim.inject_shocks(crisis_shocks)
    sim.model.options["T"] = burn_in_len
    try: df_burn = sim.run(output_vars=game["output_vars"], silent_print=True)
    except: return
    current_state = df_burn.iloc[-1]

    # C. Prompt
    state_str = ", ".join([f"{k}: {float(current_state.get(k,0)):.2f}" for k in game["output_vars"]])
    ex_key = game["allowed_shocks"][0]
    prompt = (
        f"<|im_start|>system\n"
        f"You are an Economic Policy AI. The economy is in '{scenario_name}'.\n"
        f"Goal: Stabilize variables to 0.0.\n"
        f"Output JSON format: {json.dumps({ex_key: 0.5})}\n"
        f"<|im_end|>\n"
        f"<|im_start|>user\n"
        f"Model: {game['name']}\n"
        f"State: {state_str}\n"
        f"Allowed: {json.dumps(game['allowed_shocks'])}\n"
        f"JSON Action:\n<|im_end|>\n<|im_start|>assistant\n"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    prompt_len = inputs.input_ids.shape[1]

    # D. Generate
    model.eval()
    with torch.no_grad():
        outputs = model.generate(
            **inputs, max_new_tokens=45, do_sample=True, temperature=1.2,
            num_return_sequences=GRPO_GROUP_SIZE, pad_token_id=tokenizer.eos_token_id
        )
    model.train()

    # E. Simulate & Score
    rewards = []

    print(f"\n🎮 {game['name']} | {scenario_name}")

    for i in range(GRPO_GROUP_SIZE):
        full_seq = outputs[i]
        gen_text = tokenizer.decode(full_seq[prompt_len:], skip_special_tokens=True)

        raw_act = parse_action_robust(gen_text, game["allowed_shocks"])

        # --- NEW: INACTION PENALTY ---
        if not raw_act:
            rewards.append(INACTION_PENALTY) # Massive penalty for doing nothing
            continue

        action = constrain_action(raw_act, game["allowed_shocks"], game["constraints"])

        sim_clone = sim.clone()
        sim_clone.model.options["T"] = burn_in_len + 12
        merged = merge_shocks_history(crisis_shocks, action, burn_in_len, sim.shock_names)
        sim_clone.reset_shocks()
        sim_clone.inject_shocks(merged)

        try:
            df_traj = sim_clone.run(output_vars=game["output_vars"], silent_print=True)
            loss = 0
            for v in game["output_vars"]:
                vals = df_traj[v].iloc[burn_in_len:].values
                loss += np.sum(vals**2)
            rewards.append(-loss)
        except:
            rewards.append(INACTION_PENALTY) # Penalty if sim crashes

    # F. Update
    rewards_t = torch.tensor(rewards, device=model.device, dtype=torch.float32)

    if torch.std(rewards_t) > 1e-6:
        advantages = (rewards_t - rewards_t.mean()) / (rewards_t.std() + 1e-8)

        optimizer.zero_grad()
        total_loss = 0

        for i in range(GRPO_GROUP_SIZE):
            seq = outputs[i].unsqueeze(0)
            out = model(seq)
            logits = out.logits[:, :-1, :]
            target_ids = seq[:, 1:]
            log_probs = F.log_softmax(logits, dim=-1).gather(2, target_ids.unsqueeze(-1)).squeeze(-1)

            completion_log_probs = log_probs[:, prompt_len-1:]
            seq_loss = -advantages[i] * completion_log_probs.sum() / GRPO_GROUP_SIZE
            seq_loss.backward()
            total_loss += seq_loss.item()

        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        print(f"   ✅ Loss={total_loss:.4f} | Best Reward={rewards_t.max():.2f}")
    else:
        print(f"   ⚠️ No Variance (Rewards: {rewards[0]:.2f}). Skipping.")

# --- 6. EXECUTE WITH SAFETY ---
try:
    print(f"🚀 STARTING TRAINING (Epochs: {EPOCHS})")

    for epoch in range(1, EPOCHS+1):
        print(f"\n📅 EPOCH {epoch}/{EPOCHS}")
        for game in GAME_CONFIGS:
            for scenario in game["scenarios"]:
                train_one_step(game, scenario[0], scenario[1], epoch)

        # Safe Save
        ckpt_path = f"{DRIVE_PATH}/checkpoint_epoch_{epoch}"
        model.save_pretrained(ckpt_path)
        tokenizer.save_pretrained(ckpt_path)
        print(f"💾 Saved: {ckpt_path}")

except KeyboardInterrupt:
    print("\n🛑 MANUAL STOP DETECTED. SAVING MODEL...")
    save_path = f"{DRIVE_PATH}/interrupted_save"
    model.save_pretrained(save_path)
    tokenizer.save_pretrained(save_path)
    print(f"💾 Emergency Save Complete: {save_path}")

except Exception as e:
    print(f"\n❌ CRASH DETECTED: {e}")
    traceback.print_exc()
```

```python
# ===================== CELL: ROBUST GRPO TRAINING (Fixed Drive Mount) =====================
import torch
import torch.nn.functional as F
import numpy as np
import json
import re
import copy
import yaml
import os
import sys
import traceback
import shutil
from tqdm.auto import tqdm
from google.colab import drive

from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import LoraConfig, get_peft_model, TaskType, PeftModel

# --- 1. ROBUST DRIVE MOUNTING ---
# We check if it is already mounted to avoid "Mountpoint must not already contain files" error
if os.path.exists('/content/drive/MyDrive'):
    print("✅ Drive is already mounted.")
else:
    try:
        # If the directory exists but is empty/corrupt, we clean it first
        if os.path.exists('/content/drive'):
            # Check if it's empty. If not empty and not a mount, it causes the error.
            if len(os.listdir('/content/drive')) > 0:
                print("⚠️ Cleaning /content/drive to allow mounting...")
                shutil.rmtree('/content/drive')
                os.makedirs('/content/drive')

        drive.mount('/content/drive', force_remount=True)
    except Exception as e:
        print(f"⚠️ Drive Mount Warning: {e}")
        # Proceeding anyway; if it fails here, the saving path will just be local.

# Verify path exists
DRIVE_PATH = "/content/drive/MyDrive/Economic_Agent_RL"
if not os.path.exists(DRIVE_PATH):
    try:
        os.makedirs(DRIVE_PATH)
        print(f"📂 Created directory: {DRIVE_PATH}")
    except:
        # Fallback to local if Drive fails entirely
        DRIVE_PATH = "/content/Economic_Agent_RL"
        os.makedirs(DRIVE_PATH, exist_ok=True)
        print(f"⚠️ Google Drive failed. Saving locally to: {DRIVE_PATH}")
else:
    print(f"📂 Found directory: {DRIVE_PATH}")

# --- CONFIG ---
MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"
GRPO_GROUP_SIZE = 6
EPOCHS = 100
LEARNING_RATE = 1e-5
INACTION_PENALTY = -1000.0
MAX_GRAD_NORM = 1.0
MAX_LOSS_CLIP = 5000.0     # Cap the economic crash loss

# Fix YAML
def dummy_constructor(loader, node): return None
yaml.SafeLoader.add_constructor("!MvNormal", dummy_constructor)

# --- GAME CONFIGS ---
GAME_CONFIGS = [
    {
        "name": "QPM_Full_Control",
        "model": "model.yaml",
        "output_vars": ["L_GDP_GAP", "DLA_CPI", "RS"],
        "scenarios": [
            ("Stagflation", {"SHK_DLA_CPI": [2.5]*8, "SHK_L_GDP_GAP": [-1.0]*8}),
            ("Recession", {"SHK_L_GDP_GAP": [-3.0]*8})
        ],
        "allowed_shocks": ["SHK_RS", "SHK_RR_BAR", "SHK_D4L_CPI_TAR"],
        "constraints": {"SHK_RS": (-3, 3), "SHK_RR_BAR": (-2, 2), "SHK_D4L_CPI_TAR": (-1, 1)}
    },
    {
        "name": "SW_Business_Cycles",
        "model": "sw_model.yaml",
        "output_vars": ["y", "pinf", "r"],
        "scenarios": [
            ("Productivity Crash", {"ea": [-5.0]*8}),
            ("Cost Push", {"epinf": [3.0]*8})
        ],
        "allowed_shocks": ["em", "eg", "ea"],
        "constraints": {"em": (-3.0, 3.0), "eg": (-3.0, 3.0), "ea": (0, 2.0)}
    },
    {
        "name": "GSW_Pandemic_Sim",
        "model": "gsw_model.yaml",
        "output_vars": ["y", "unempl", "pinf"],
        "scenarios": [("COVID Lockdown", {"els": [10.0]*8})],
        "allowed_shocks": ["em", "eg", "ew"],
        "constraints": {"em": (-2, 2), "eg": (0, 8.0), "ew": (-2, 2)}
    },
    {
        "name": "Ireland_NK_Tech",
        "model": "Ireland2004.yaml",
        "output_vars": ["y", "pie", "r"],
        "scenarios": [("Tech Collapse", {"epsa": [-3.0]*8})],
        "allowed_shocks": ["epsr", "epsz"],
        "constraints": {"epsr": (-2, 2), "epsz": (-1, 1)}
    },
    {
        "name": "MVF_US_Gap",
        "model": "MVF_US.yaml",
        "output_vars": ["Y", "PIE", "UNR_GAP"],
        "scenarios": [("Overheating", {"RES_Y": [3.0]*8, "RES_PIE": [1.5]*8})],
        "allowed_shocks": ["RES_Y"],
        "constraints": {"RES_Y": (-3, 3)}
    }
]

# --- IMPROVED PARSER ---
def parse_action_robust(text: str, allowed_keys: list) -> dict:
    text = text.strip()
    extracted = {}

    # Strategy 1: Look for explicit Key: Value (e.g., "em": 0.5)
    for key in allowed_keys:
        pattern = r'["\']?' + re.escape(key) + r'["\']?\s*[:=]\s*([-+]?\d*\.?\d+)'
        match = re.search(pattern, text, re.IGNORECASE)
        if match: extracted[key] = float(match.group(1))

    # Strategy 2: Look for LLM "Action/Value" structure
    if not extracted:
        try:
            # Matches: "Action": "SHK_RS", "Value": -0.1
            act_match = re.search(r'["\']?action["\']?\s*[:=]\s*["\']?(\w+)["\']?', text, re.IGNORECASE)
            val_match = re.search(r'["\']?value["\']?\s*[:=]\s*([-+]?\d*\.?\d+)', text, re.IGNORECASE)

            if act_match and val_match:
                k = act_match.group(1)
                v = float(val_match.group(1))
                if k in allowed_keys:
                    extracted[k] = v
        except: pass

    # Strategy 3: JSON Fallback
    if not extracted:
        try:
            match = re.search(r"\{.*\}", text, re.DOTALL)
            if match:
                clean_json = re.sub(r",\s*}", "}", match.group(0))
                data = json.loads(clean_json)
                if isinstance(data, dict):
                    for k, v in data.items():
                        if k in allowed_keys: extracted[k] = float(v)
        except: pass

    return extracted

def constrain_action(action, allowed, constraints):
    out = {}
    for k, v in action.items():
        if k in allowed:
            try:
                val = float(v)
                mn, mx = constraints.get(k, (-2.5, 2.5))
                out[k] = max(mn, min(mx, val))
            except: pass
    return out

def merge_shocks_history(crisis_shocks, agent_action, burn_in_len, shock_names):
    final_dict = {}
    for k, v_list in crisis_shocks.items():
        final_dict[k] = [(t, val) for t, val in enumerate(v_list)]
    for k, val in agent_action.items():
        if k in shock_names:
            if k not in final_dict: final_dict[k] = []
            final_dict[k].append((burn_in_len, float(val)))
            for p in range(1, 4):
                 final_dict[k].append((burn_in_len + p, float(val) * (0.8**p)))
    return final_dict

# --- MODEL LOADING ---
last_ckpt = f"{DRIVE_PATH}/checkpoint_latest"
if os.path.exists(last_ckpt):
    print(f"🔄 Resuming from: {last_ckpt}")
    tokenizer = AutoTokenizer.from_pretrained(last_ckpt, trust_remote_code=True)
    base_model = AutoModelForCausalLM.from_pretrained(MODEL_ID, torch_dtype=torch.float16, device_map="auto", trust_remote_code=True)
    model = PeftModel.from_pretrained(base_model, last_ckpt, is_trainable=True)
else:
    print("⚙️ Initializing New Model (QLoRA)...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, trust_remote_code=True)
    tokenizer.pad_token = tokenizer.eos_token
    base_model = AutoModelForCausalLM.from_pretrained(MODEL_ID, torch_dtype=torch.float16, device_map="auto", trust_remote_code=True)
    peft_config = LoraConfig(task_type=TaskType.CAUSAL_LM, r=16, lora_alpha=32, lora_dropout=0.05, target_modules=["q_proj", "k_proj", "v_proj", "o_proj"])
    model = get_peft_model(base_model, peft_config)

model.config.use_cache = False
optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)
print("✅ Model Ready.")

# --- TRAIN STEP ---
def train_one_step(game, scenario_name, crisis_shocks):
    try:
        path = game["model"]
        if not os.path.exists(path) and os.path.exists(f"/content/{path}"): path = f"/content/{path}"
        sim = GenericDSGESimulator(path, periods=25, silence_snowdrop=True)
        if "shock_values" in sim.model.options: del sim.model.options["shock_values"]
        parser = RobustModelParser(path)
    except: return

    # Burn-in
    burn_in_len = 8
    sim.reset_shocks()
    sim.inject_shocks(crisis_shocks)
    sim.model.options["T"] = burn_in_len
    try:
        df_burn = sim.run(output_vars=game["output_vars"], silent_print=True)
        current_state = df_burn.iloc[-1]
    except: return

    state_str = ", ".join([f"{k}: {float(current_state.get(k,0)):.2f}" for k in game["output_vars"]])
    ex_key = game["allowed_shocks"][0]

    # --- UPDATED PROMPT: More specific instructions ---
    prompt = (
        f"<|im_start|>system\n"
        f"You are an Economic Policy AI. The economy is in '{scenario_name}'.\n"
        f"Goal: Stabilize variables to 0.0. \n"
        f"Strict JSON Output: {{\"{ex_key}\": 0.5}}\n"
        f"Do not use nested keys like 'action' or 'value'.\n"
        f"<|im_end|>\n"
        f"<|im_start|>user\n"
        f"Model: {game['name']}\n"
        f"State: {state_str}\n"
        f"Allowed: {json.dumps(game['allowed_shocks'])}\n"
        f"JSON Action:\n<|im_end|>\n<|im_start|>assistant\n"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    prompt_len = inputs.input_ids.shape[1]

    # Generate Group
    model.eval()
    with torch.no_grad():
        outputs = model.generate(
            **inputs, max_new_tokens=45, do_sample=True, temperature=1.2, top_p=0.9,
            num_return_sequences=GRPO_GROUP_SIZE, pad_token_id=tokenizer.eos_token_id
        )
    model.train()

    rewards = []
    actions_text = []

    for i in range(GRPO_GROUP_SIZE):
        full_seq = outputs[i]
        gen_text = tokenizer.decode(full_seq[prompt_len:], skip_special_tokens=True)
        actions_text.append(gen_text)

        # New Parser used here
        raw_act = parse_action_robust(gen_text, game["allowed_shocks"])

        if not raw_act:
            rewards.append(INACTION_PENALTY)
            continue

        action = constrain_action(raw_act, game["allowed_shocks"], game["constraints"])

        sim_clone = sim.clone()
        sim_clone.model.options["T"] = burn_in_len + 12
        merged = merge_shocks_history(crisis_shocks, action, burn_in_len, sim.shock_names)
        sim_clone.reset_shocks()
        sim_clone.inject_shocks(merged)

        try:
            df_traj = sim_clone.run(output_vars=game["output_vars"], silent_print=True)
            loss = 0
            for v in game["output_vars"]:
                vals = df_traj[v].iloc[burn_in_len:].values
                loss += np.sum(vals**2)

            # --- CLIPPING: Prevent Infinite Losses ---
            loss = min(loss, MAX_LOSS_CLIP)
            rewards.append(-loss)
        except:
            rewards.append(INACTION_PENALTY)

    rewards_t = torch.tensor(rewards, device=model.device, dtype=torch.float32)

    # Logging
    best_idx = torch.argmax(rewards_t).item()
    best_val = rewards_t[best_idx].item()
    clean_act = actions_text[best_idx].replace("\n", " ")[:60] # Truncate long strings

    if torch.std(rewards_t) > 1e-6:
        advantages = (rewards_t - rewards_t.mean()) / (rewards_t.std() + 1e-8)

        optimizer.zero_grad()
        total_loss = 0

        for i in range(GRPO_GROUP_SIZE):
            seq = outputs[i].unsqueeze(0)
            out = model(seq)
            logits = out.logits[:, :-1, :]
            target_ids = seq[:, 1:]
            log_probs = F.log_softmax(logits, dim=-1).gather(2, target_ids.unsqueeze(-1)).squeeze(-1)

            completion_log_probs = log_probs[:, prompt_len-1:]
            seq_loss = -(advantages[i] * completion_log_probs.sum()) / GRPO_GROUP_SIZE
            seq_loss.backward()
            total_loss += seq_loss.item()

        torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
        optimizer.step()
        return total_loss, best_val, clean_act
    return 0.0, best_val, clean_act

# ===================== CELL: EXECUTION WITH PERMANENT LOGS =====================

# ... (Keep all your imports and helper functions above this same as before) ...

# --- MAIN LOOP ---
try:
    print(f"🚀 STARTING TRAINING (Epochs: {EPOCHS})")
    print(f"📂 Saving to: {DRIVE_PATH} (Check your Google Drive folder)")

    # 1. Epoch Progress Bar
    epoch_bar = tqdm(range(1, EPOCHS+1), unit="epoch", desc="Overall Progress")

    for epoch in epoch_bar:
        epoch_loss = 0
        step_count = 0

        tqdm.write(f"\n📅 === STARTING EPOCH {epoch} ===")

        # 2. Flatten tasks so we can iterate nicely
        tasks = []
        for game in GAME_CONFIGS:
            for scenario in game["scenarios"]:
                tasks.append((game, scenario))

        # 3. Scenario Progress Bar
        # We set leave=False so this bar clears after the epoch is done, leaving clean logs
        scenario_bar = tqdm(tasks, leave=False, desc=f"Ep {epoch} Scenarios")

        for game, (scen_name, scen_shocks) in scenario_bar:
            # Update the description to show what we are working on
            scenario_bar.set_description(f"Ep {epoch} | {game['name']} | {scen_name}")

            # Run the training step
            loss, best_r, best_act = train_one_step(game, scen_name, scen_shocks)

            # --- THE FIX: Use tqdm.write for permanent logging ---
            # This prints ABOVE the progress bar and stays there
            log_msg = (
                f"   🎮 {game['name'][:15]}.. | {scen_name[:15]}.. "
                f"| 🏆 Rew: {best_r:.1f} | 🎬 Act: {best_act}"
            )
            tqdm.write(log_msg)

            if loss != 0:
                epoch_loss += loss
                step_count += 1

        avg_loss = epoch_loss / step_count if step_count > 0 else 0
        tqdm.write(f"✅ Epoch {epoch} Done | Avg Loss: {avg_loss:.2f}\n")

        # --- SAVE CHECKPOINT ---
        # Save specific epoch
        ckpt_path = f"{DRIVE_PATH}/checkpoint_epoch_{epoch}"
        model.save_pretrained(ckpt_path)
        tokenizer.save_pretrained(ckpt_path)

        # Save 'latest' for resuming
        latest_path = f"{DRIVE_PATH}/checkpoint_latest"
        model.save_pretrained(latest_path)
        tokenizer.save_pretrained(latest_path)

        tqdm.write(f"💾 Saved Checkpoint to Drive: .../checkpoint_epoch_{epoch}")

    print("✅ Training Complete.")

except KeyboardInterrupt:
    print("\n🛑 Manual Stop. Saving...")
    model.save_pretrained(f"{DRIVE_PATH}/interrupted")
    tokenizer.save_pretrained(f"{DRIVE_PATH}/interrupted")
    print("💾 Emergency Save Complete.")

except Exception as e:
    print(f"\n❌ Error: {e}")
    traceback.print_exc()
```
