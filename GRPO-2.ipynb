````python
# RobustModelParser.py
import yaml
import re
from pathlib import Path

class RobustModelParser:
    """
    Extracts variables, shocks, parameters, calibration, equations,
    and labels from YAML and comment-style lines.
    """

    def __init__(self, file_path):
        self.file_path = Path(file_path)
        self.raw_text = self.file_path.read_text()
        self.yaml_obj = yaml.safe_load(self.raw_text)

        self.comment_labels = self._extract_comment_labels()
        self.yaml_labels = self.yaml_obj.get("labels", {}) or {}
        self.labels = {**self.comment_labels, **self.yaml_labels}

    def _extract_comment_labels(self):
        pattern = r"#\s*(.*?)\s{2,}([A-Za-z0-9_]+)\s*$"
        matches = re.findall(pattern, self.raw_text, flags=re.MULTILINE)
        return {var.strip(): desc.strip() for desc, var in matches}

    def symbols(self):
        return self.yaml_obj.get("symbols", {}) or {}

    def variables(self):
        return self.symbols().get("variables", []) or []

    def measurement_variables(self):
        return self.symbols().get("measurement_variables", []) or []

    def shocks(self):
        return self.symbols().get("shocks", []) or []

    def measurement_shocks(self):
        return self.symbols().get("measurement_shocks", []) or []

    def parameters(self):
        return self.symbols().get("parameters", []) or []

    def calibration(self):
        return self.yaml_obj.get("calibration", {}) or {}

    def equations(self):
        return (
            self.yaml_obj.get("equations", []) or [],
            self.yaml_obj.get("measurement_equations", []) or []
        )

    def build_llm_context(self):
        eqs, meqs = self.equations()
        ctx = f"MODEL CONTEXT — {self.file_path.name}\n\n"

        def add(title, arr):
            nonlocal ctx
            ctx += f"--- {title} ---\n"
            for x in arr:
                label = self.labels.get(x, "(no label)")
                ctx += f"  - {x}: {label}\n"
            ctx += "\n"

        add("VARIABLES", self.variables())
        add("MEASUREMENT VARIABLES", self.measurement_variables())
        add("PARAMETERS", self.parameters())
        add("SHOCKS", self.shocks())
        add("MEASUREMENT SHOCKS", self.measurement_shocks())

        ctx += "--- CALIBRATION ---\n"
        for k, v in self.calibration().items():
            ctx += f"  - {k}: {v}\n"

        ctx += "\n--- TRANSITION EQUATIONS ---\n"
        for e in eqs:
            ctx += f"  - {e}\n"

        ctx += "\n--- MEASUREMENT EQUATIONS ---\n"
        for e in meqs:
            ctx += f"  - {e}\n"

        return ctx
````

````python
# generic_dsge_simulator.py
import copy
import io
import sys
from typing import Optional, List, Union, Dict, Any

import numpy as np
import pandas as pd

# Snowdrop imports (assumes snowdrop is on PYTHONPATH)
from snowdrop.src.driver import importModel, run
from snowdrop.src.model.util import setValues

# ---------------------------
# Generic DSGE Simulator (final)
# ---------------------------
class GenericDSGESimulator:
    """
    Robust, model-agnostic wrapper around Snowdrop models.

    Features:
      - inject_shocks supports: scalar, list, list-of-(t,val), pandas.Series
      - update_parameters supports scalar and time-varying calibrated arrays (via setValues fallback)
      - reset_shocks to zero the shock matrix for current horizon
      - clone() to create an identical simulator preserving calibration/options (deep copy)
      - auto solver selection based on model.options["linear"]
      - generate_irf() to compute IRF = shocked - baseline using only the public API
      - rl_step() safe single-step interface for RL
    """

    def __init__(self, model_path: str, periods: int = 40, silence_snowdrop: bool = True):
        self.model_path = str(model_path)
        self.periods = int(periods)

        # optionally suppress Snowdrop import-time prints
        _stdout = sys.stdout
        if silence_snowdrop:
            sys.stdout = io.StringIO()
        try:
            self.model = importModel(self.model_path)
        finally:
            if silence_snowdrop:
                sys.stdout = _stdout

        # ensure T/periods is set (Snowdrop uses options["T"])
        self.model.options["T"] = self.periods
        self.model.options.setdefault("range", None)

        # cache symbol lists
        s = getattr(self.model, "symbols", {})
        self.vars = list(s.get("variables", []))
        self.shock_names = list(s.get("shocks", []))
        self.param_names = list(s.get("parameters", []))

    # ---------------------------
    # Helper: choose default solver based on model linearity
    # ---------------------------
    def _choose_solver(self, solver: Optional[str]) -> str:
        if solver is not None:
            return solver

        # Snowdrop convention: model.options['linear'] frequently exists.
        is_linear = self.model.options.get("linear", True)
        if is_linear:
            # Linear solver default
            return "BinderPesaran"
        else:
            return "LBJ"

    # ---------------------------
    # Shock injection (robust)
    # ---------------------------
    def inject_shocks(self, shock_dict: Dict[str, Any], reset: bool = True):
        """
        Inject shocks into the model.

        shock_dict formats accepted:
          - scalar: {"epsr": 0.1}
          - list:   {"epsr": [0.1, 0.0, ...]}
          - tuples: {"em": [(0,0.02),(3,-0.01)]}
          - pandas.Series indexed by 0..T-1 (or with datetime index, but we use positional)
        """
        import pandas as _pd

        T = int(self.model.options.get("T", self.periods))
        nshk = len(self.shock_names)
        S = np.zeros((T, nshk))

        for shk, values in shock_dict.items():
            if shk not in self.shock_names:
                print(f"⚠️ Warning: unknown shock '{shk}' — ignoring.")
                continue
            j = self.shock_names.index(shk)

            # scalar
            if np.isscalar(values):
                S[0, j] = float(values)
                continue

            # pandas Series
            if isinstance(values, _pd.Series):
                for idx, val in enumerate(values.iloc[:T]):
                    S[idx, j] = float(val)
                continue

            # list of tuples (t, val)
            if isinstance(values, list) and len(values) > 0 and isinstance(values[0], tuple):
                for (t, val) in values:
                    ti = int(t)
                    if 0 <= ti < T:
                        S[ti, j] = float(val)
                continue

            # list of values (path)
            if isinstance(values, list):
                # truncate/pad
                for i, val in enumerate(values[:T]):
                    S[i, j] = float(val)
                continue

            raise ValueError(f"Invalid shock format for {shk}: {values}")

        # write back to model options and calibration if present
        self.model.options["shock_values"] = S
        try:
            # some models expect calibration["shocks"]
            self.model.calibration["shocks"] = S
        except Exception:
            pass

    # ---------------------------
    # Reset shocks to zero for the current T horizon
    # ---------------------------
    def reset_shocks(self):
        T = int(self.model.options.get("T", self.periods))
        nshk = len(self.shock_names)
        zeros = np.zeros((T, nshk))
        self.model.options["shock_values"] = zeros
        try:
            self.model.calibration["shocks"] = zeros
        except Exception:
            pass

    # ---------------------------
    # Update parameters (robust)
    # ---------------------------
    def update_parameters(self, param_dict: Dict[str, Any], start: Optional[Any] = None):
        """
        Uses setValues for robust mapping when available, otherwise updates direct calibration.
        """
        if not param_dict:
            return

        # Try using Snowdrop's setValues to handle arrays/time-varying parameters
        try:
            curr = self.model.calibration.get("parameters", None)
            new_calib, _, _ = setValues(
                model=self.model,
                d=param_dict,
                names=self.param_names,
                values=curr,
                start=start,
                isShock=False
            )
            self.model.calibration["parameters"] = new_calib
            return
        except Exception:
            # fallback: direct overwrite where possible
            param_values = self.model.calibration.get("parameters", None)
            if param_values is None:
                raise RuntimeError("Model calibration['parameters'] not found and setValues failed.")
            for name, new_val in param_dict.items():
                if name in self.param_names:
                    idx = self.param_names.index(name)
                    val = param_values[idx]
                    if np.isscalar(val):
                        param_values[idx] = new_val
                    else:
                        # assume array-like
                        param_values[idx][:] = new_val
            self.model.calibration["parameters"] = param_values

    # ---------------------------
    # Run simulation (suppresses Snowdrop prints and returns DataFrame)
    # ---------------------------
    def run(self, solver: Optional[str] = None, output_vars: Optional[List[str]] = None, silent_print: bool = False) -> pd.DataFrame:
        """
        Run the model and return a DataFrame with result variables (or subset output_vars).
        This method suppresses Snowdrop's internal print output.
        """
        solver_to_use = self._choose_solver(solver)

        # suppress prints
        _stdout = sys.stdout
        sys.stdout = io.StringIO()
        try:
            # Some Snowdrop versions accept "output_variables" kwarg, but to be robust we'll call run normally
            y, rng = run(
                model=self.model,
                Solver=solver_to_use,
                Plot=False,
                Output=False
            )
        finally:
            sys.stdout = _stdout

        # ensure y is an ndarray and rng is indexable
        rows = min(len(rng), getattr(y, "shape", [0])[0])
        df = pd.DataFrame(y[:rows], index=rng[:rows], columns=self.vars)

        # if user requested subset of vars, validate and return subset
        if output_vars:
            missing = [v for v in output_vars if v not in df.columns]
            if missing:
                raise ValueError(f"Requested output_vars not present in model: {missing}")
            df_print = df[output_vars]
        else:
            df_print = df

        # print only when not silent_print
        if not silent_print:
            print("\n=== SIMULATION OUTPUT (subset) ===")
            print(df_print.head(12))
            print("\n=== STATISTICS (subset) ===")
            print(df_print.describe())

        return df_print

    # ---------------------------
    # Clone current simulator (deep copy of model state) — returns new GenericDSGESimulator
    # ---------------------------
    def clone(self):
        """
        Create a fresh simulator instance with the SAME model_path and periods,
        then deep-copy calibration and options so the clone has the same current state.
        """
        new_sim = GenericDSGESimulator(self.model_path, periods=self.periods, silence_snowdrop=True)
        # deep copy current model's calibration and options into the new model
        new_sim.model.calibration = copy.deepcopy(getattr(self.model, "calibration", {}))
        new_sim.model.options = copy.deepcopy(getattr(self.model, "options", {}))

        # ensure new_sim caches are updated (symbols same)
        s = getattr(new_sim.model, "symbols", {})
        new_sim.vars = list(s.get("variables", []))
        new_sim.shock_names = list(s.get("shocks", []))
        new_sim.param_names = list(s.get("parameters", []))

        return new_sim

    # ---------------------------
    # IRF generator (model-agnostic)
    # ---------------------------
    def generate_irf(
        self,
        shock: str,
        size: float = 1.0,
        horizon: Optional[int] = None,
        vars: Optional[List[str]] = None,
        shock_at: int = 0,
        solver: Optional[str] = None,
        use_shock_std: bool = False,
        silent_print: bool = False
    ) -> pd.DataFrame:
        """
        Compute IRF = (shocked simulation) - (baseline simulation).

        Args:
          shock: name of shock in model.symbols['shocks']
          size: magnitude of shock
          horizon: #periods for IRF (defaults to self.periods)
          vars: list of variable names to return (subset). If None returns all.
          shock_at: integer period index where shock is applied (default t=0)
          solver: optional override solver
          use_shock_std: if True, look up calibration key 'std_<shock>' and multiply size by it

        Returns:
          DataFrame indexed by simulation dates, columns = requested variables; values = shocked - baseline
        """
        if horizon is None:
            horizon = self.periods
        horizon = int(horizon)

        if shock not in self.shock_names:
            raise ValueError(f"Shock '{shock}' not in model.shocks: {self.shock_names}")

        # optionally scale by calibrated std
        mag = float(size)
        if use_shock_std:
            std_key = f"std_{shock}"
            std_val = None
            calib = getattr(self.model, "calibration", {})
            if isinstance(calib, dict) and std_key in calib:
                std_val = calib[std_key]
            if std_val is None:
                print(f"⚠️ WARNING: std_{shock} not found in calibration; proceeding with raw size.")
            else:
                mag = mag * float(std_val)

        # --------------------------
        # Run baseline (clone and reset shocks)
        # --------------------------
        base_sim = self.clone()
        base_sim.model.options["T"] = horizon
        base_sim.reset_shocks()
        baseline_df = base_sim.run(solver=solver, output_vars=vars, silent_print=silent_print)

        # --------------------------
        # Run shocked sim (clone current state so param changes are preserved)
        # --------------------------
        shocked_sim = self.clone()
        shocked_sim.model.options["T"] = horizon
        shocked_sim.reset_shocks()
        shocked_sim.inject_shocks({shock: [(shock_at, mag)]})
        shocked_df = shocked_sim.run(solver=solver, output_vars=vars, silent_print=silent_print)

        # align indices/columns
        baseline_df, shocked_df = baseline_df.align(shocked_df, join="outer", axis=0)
        irf_df = shocked_df.fillna(0) - baseline_df.fillna(0)
        irf_df = irf_df.iloc[:horizon]

        if not silent_print:
            print(f"\n=== IRF for shock '{shock}' (size={mag}) ===")
            print(irf_df.head(12))

        return irf_df

    # ---------------------------
    # RL STEP (safe version)
    # ---------------------------
    def rl_step(self, action_shocks: Dict[str, Any], param_updates: Optional[Dict[str, Any]] = None, output_vars: Optional[List[str]] = None, solver: Optional[str] = None):
        """
        Performs a single-step transition for RL agents.

        This:
            - clones the simulator (so original T is preserved)
            - runs a T=1 simulation with the given shock and parameters
            - returns the single-step resulting state (last row)
        """
        temp = self.clone()
        temp.model.options["T"] = 1

        # apply RL-driven parameter updates (policy changes) to the temp sim
        if param_updates:
            temp.update_parameters(param_updates)

        # place action shocks at t=0
        temp.reset_shocks()
        temp.inject_shocks(action_shocks)

        # run one-step model and return last row
        df = temp.run(solver=solver, output_vars=output_vars, silent_print=True)
        # df might be empty or single-row; return a Series or empty frame accordingly
        if df.shape[0] == 0:
            return df
        return df.iloc[-1]
````

````python
# ===================== CELL 1: SLEDGEHAMMER CONFIG =====================
import numpy as np
import json
import re
import copy
import os

# 1. Semantic Maps (Unchanged)
INTENSITY_MAP = {
    "MILD": 0.25, "LOW": 0.25, "SMALL": 0.25,
    "MODERATE": 0.50, "MEDIUM": 0.50, "STANDARD": 0.50,
    "AGGRESSIVE": 1.0, "HIGH": 1.0, "STRONG": 1.0, # Increased scalar
    "EMERGENCY": 2.0, "EXTREME": 2.0, "MAX": 2.0   # Increased scalar
}

DIRECTION_MAP = {
    "TIGHTEN": 1.0, "HIKE": 1.0, "CONTRACT": 1.0, "INCREASE": 1.0,
    "EASE": -1.0, "CUT": -1.0, "EXPAND": -1.0, "DECREASE": -1.0, "LOWER": -1.0,
    "HOLD": 0.0, "STABILIZE": 0.0
}

# 2. AMPLIFIED MODEL WORLD (The "Sledgehammer")
MODEL_WORLD = {
    "QPM_Full_Control": {
        "path": "model.yaml",
        "labels": {"DLA_CPI": "Inflation", "L_GDP_GAP": "Output Gap"},
        "targets": ["DLA_CPI", "L_GDP_GAP"],
        "mappings": {
            "MONETARY": {"var": "SHK_RS", "flip": False},
            "FISCAL": {"var": "SHK_L_GDP_GAP", "flip": True}
        },
        # FIX: Massive, sustained shocks
        "scenarios": {
            "HYPERINFLATION": {"SHK_DLA_CPI": [5.0]*12}, # 12 Quarters of 5% shocks
            "DEPRESSION": {"SHK_L_GDP_GAP": [-5.0]*12}   # 12 Quarters of -5% shocks
        }
    },
    "SW_Business_Cycles": {
        "path": "sw_model.yaml",
        "labels": {"pinf": "Inflation", "y": "Output"},
        "targets": ["pinf", "y"],
        "mappings": {
            "MONETARY": {"var": "em", "flip": False},
            "FISCAL": {"var": "eg", "flip": False},
            "TECH": {"var": "ea", "flip": False}
        },
        "scenarios": {
            "SUPPLY_COLLAPSE": {"epinf": [6.0]*12}, # Massive Cost Push
            "TECH_CRASH": {"ea": [-6.0]*12}        # Productivity vanishes
        }
    },
    # GSW Removed temporarily to focus learning on QPM/SW first
}

# 3. Robust Parser (Unchanged)
def translate_semantic_action(text, model_key):
    config = MODEL_WORLD[model_key]
    text = text.upper().replace("```JSON", "").replace("```", "").strip()
    match = re.search(r"\{.*\}", text, re.DOTALL)
    data_str = match.group(0) if match else text

    selected_map = None
    for pol_key, map_data in config["mappings"].items():
        if pol_key in data_str:
            selected_map = map_data
            break
    if not selected_map: return {}

    scalar = 0.5
    for word, val in INTENSITY_MAP.items():
        if re.search(r"\b" + re.escape(word) + r"\b", data_str):
            scalar = val; break

    sign = 0.0
    for word, s_val in DIRECTION_MAP.items():
        if re.search(r"\b" + re.escape(word) + r"\b", data_str):
            sign = s_val; break

    if selected_map["flip"]: sign *= -1.0
    return {selected_map["var"]: sign * scalar}
````

````python
# ===================== CELL 2: GYM WITH INACTION PENALTY =====================


class HinterlangGym:
    def __init__(self, model_key, sim_class):
        self.conf = MODEL_WORLD[model_key]
        p = self.conf["path"]
        if not os.path.exists(p) and os.path.exists(f"/content/{p}"): p = f"/content/{p}"

        self.sim = sim_class(p, periods=50, silence_snowdrop=True)
        self.horizon = 12
        self.burn_in = 8

    def _calculate_loss(self, df):
        if len(df) < self.burn_in + self.horizon: return 1e6
        loss = 0.0
        for target in self.conf["targets"]:
            traj = df[target].iloc[self.burn_in : self.burn_in+self.horizon].values
            w = 2.0 if "CPI" in target or "pinf" in target else 1.0
            loss += np.sum((traj**2) * w)
        return loss

    def reset(self, scenario_name):
        self.current_scenario = self.conf["scenarios"][scenario_name]

        base_sim = self.sim.clone()
        base_sim.reset_shocks()
        base_sim.inject_shocks(self.current_scenario)
        try:
            df = base_sim.run(silent_print=True)
            self.baseline_loss = self._calculate_loss(df)

            last_row = df.iloc[self.burn_in]
            state = {}
            for k in self.conf["targets"]:
                label = self.conf["labels"].get(k, k)
                state[label] = round(float(last_row.get(k, 0)), 2)

            return state, self.baseline_loss
        except:
            return {}, 1e6

    def step(self, agent_shocks):
        # 1. Check for Inaction (Empty Dict or Zero Values)
        is_inaction = False
        if not agent_shocks:
            is_inaction = True
        elif all(v == 0 for v in agent_shocks.values()):
            is_inaction = True

        sim = self.sim.clone()
        merged = copy.deepcopy(self.current_scenario)

        # 2. Stronger Persistence (Decay slower: 0.9)
        for shk, val in agent_shocks.items():
            if shk not in merged: merged[shk] = [0.0]*60
            elif len(merged[shk]) < 60: merged[shk] += [0.0]*(60-len(merged[shk]))

            curr = val
            for t in range(self.burn_in, self.burn_in + 6): # Extend influence to 6 quarters
                if t < 60:
                    merged[shk][t] += curr
                    curr *= 0.9 # Slower decay = Stronger Policy

        sim.reset_shocks()
        sim.inject_shocks(merged)

        try:
            df = sim.run(silent_print=True)
            agent_loss = self._calculate_loss(df)

            # 3. Reward Logic
            delta = self.baseline_loss - agent_loss

            # Base Reward (Improvement)
            reward = np.tanh(delta / 500.0) * 10.0 # Divisor increased for larger shocks

            # Inaction Penalty: If house is burning (Base Loss > 1000) and agent did nothing
            if is_inaction and self.baseline_loss > 1000:
                reward -= 5.0 # Force action

            return reward, agent_loss, is_inaction
        except:
            return -10.0, 1e6, False
````

````python
# ===================== CELL 3: DEBUG TRAINING LOOP =====================
import torch
import torch.nn.functional as F
from tqdm.auto import tqdm
import random

def train_debug_grpo(model, tokenizer, sim_class):

    EPOCHS = 5
    GROUP_SIZE = 4
    LR = 5e-6
    MAX_GRAD_NORM = 1.0

    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)

    print(f"🚀 STARTING DEBUG TRAINING (HIGH SHOCK MAGNITUDE)")

    for epoch in range(EPOCHS):
        print(f"\n📅 === EPOCH {epoch+1} ===")

        tasks = []
        for m_key, m_val in MODEL_WORLD.items():
            for s_key in m_val["scenarios"]:
                tasks.append((m_key, s_key))
        random.shuffle(tasks)

        bar = tqdm(tasks, desc=f"Ep {epoch+1}")

        for m_key, s_key in bar:
            gym = HinterlangGym(m_key, sim_class)
            state, base_loss = gym.reset(s_key)

            # Skip if baseline is too small (Equilibrium Trap)
            if base_loss < 100:
                # print(f"Skipping {s_key} (Loss {base_loss:.0f} too low)")
                continue

            menu_pol = list(MODEL_WORLD[m_key]["mappings"].keys())

            prompt = (
                f"<|im_start|>system\n"
                f"You are a Central Bank AI. Crisis: {s_key}.\n"
                f"Objective: Minimize cumulative loss (Baseline: {base_loss:,.0f}).\n"
                f"Policies: {menu_pol}. Directions: [TIGHTEN, EASE, HOLD].\n"
                f"Intensities: [MILD, MODERATE, AGGRESSIVE, EMERGENCY].\n"
                f"Format: JSON.\n"
                f"<|im_end|>\n"
                f"<|im_start|>user\n"
                f"Dashboard: {json.dumps(state)}\n"
                f"Action:\n<|im_end|>\n<|im_start|>assistant\n"
            )

            inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
            p_len = inputs.input_ids.shape[1]

            model.eval()
            with torch.no_grad():
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=50,
                    do_sample=True,
                    temperature=1.5, # High exploration
                    top_p=0.90,
                    num_return_sequences=GROUP_SIZE,
                    pad_token_id=tokenizer.eos_token_id
                )
            model.train()

            rewards = []

            # DEBUG: Print one example per batch
            debug_info = None

            for i in range(GROUP_SIZE):
                text = tokenizer.decode(outputs[i][p_len:], skip_special_tokens=True)
                shocks = translate_semantic_action(text, m_key)

                # Check for Parse Fail
                parse_fail = False
                if not shocks and "HOLD" not in text.upper():
                    parse_fail = True

                if parse_fail:
                    rewards.append(-5.0)
                    status = "PARSE_FAIL"
                else:
                    r, agent_loss, inactive = gym.step(shocks)
                    rewards.append(r)
                    if inactive: status = "INACTION"
                    elif r > 0: status = "IMPROVED"
                    else: status = "WORSENED"

                if i == 0: # Save first item for debug print
                    debug_info = f"[{status}] Act: {str(shocks)} | R: {rewards[-1]:.2f} | Text: {text[:30]}"

            # Print Debug info to console (above progress bar)
            # tqdm.write(debug_info)

            rt = torch.tensor(rewards, device=model.device, dtype=torch.float32)
            if torch.std(rt) < 1e-6:
                adv = torch.zeros_like(rt)
            else:
                adv = (rt - rt.mean()) / (rt.std() + 1e-8)

            optimizer.zero_grad()
            for i in range(GROUP_SIZE):
                seq = outputs[i].unsqueeze(0)
                out = model(seq)
                logits = out.logits[:, :-1, :]
                targets = seq[:, 1:]
                log_probs = F.log_softmax(logits, dim=-1).gather(2, targets.unsqueeze(-1)).squeeze(-1)
                mask = torch.zeros_like(log_probs)
                mask[:, p_len-1:] = 1.0
                loss = -(adv[i] * (log_probs * mask).sum()) / GROUP_SIZE
                loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
            optimizer.step()

            bar.set_description(f"Ep{epoch+1}|{m_key[:3]}|Best:{max(rewards):.1f}")

        ckpt_path = f"/content/drive/MyDrive/Economic_Agent_RL/Checkpoints/epoch_{epoch+1}"
        model.save_pretrained(ckpt_path)
        print(f"✅ Epoch {epoch+1} Saved. Example interaction:\n   {debug_info}")
train_debug_grpo(model, tokenizer, GenericDSGESimulator)
````

````python
# ===================== CELL 3: CALIBRATED PRODUCTION TRAINING =====================
import torch
import torch.nn.functional as F
from tqdm.auto import tqdm
import random
import numpy as np

# --- FIX 1: MICRO-SCALARS (The Scalpel) ---
# We reduced these values significantly to prevent "exploding" the simulator
INTENSITY_SCALARS = {
    "MILD": 0.05,       "LOW": 0.05,      "SMALL": 0.05,
    "MODERATE": 0.10,   "MEDIUM": 0.10,   "STANDARD": 0.10,
    "AGGRESSIVE": 0.20, "HIGH": 0.20,     "STRONG": 0.20,
    "EMERGENCY": 0.30,  "EXTREME": 0.30,  "MAX": 0.30
}

# Keep Directions the same
DIRECTION_SIGNS = {
    "TIGHTEN": 1.0, "HIKE": 1.0, "CONTRACT": 1.0, "INCREASE": 1.0, "RAISE": 1.0,
    "EASE": -1.0, "CUT": -1.0, "EXPAND": -1.0, "LOWER": -1.0, "DECREASE": -1.0,
    "HOLD": 0.0, "WAIT": 0.0
}

def train_production_grpo_calibrated(model, tokenizer, sim_class):

    # --- HYPERPARAMETERS ---
    EPOCHS = 5
    GROUP_SIZE = 4
    LR = 5e-6
    MAX_GRAD_NORM = 1.0

    # --- FIX 2: ULTRA-HAWK WEIGHTS ---
    # We ignore GDP variance because it causes the agent to freeze
    W_CPI = 3.0
    W_GDP = 0.1

    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)

    print(f"🚀 STARTING CALIBRATED TRAINING (Micro-Actions + Hawk Weights)")

    for epoch in range(EPOCHS):
        print(f"\n📅 === EPOCH {epoch+1} ===")
        epoch_rewards = []

        tasks = []
        for m_key, m_val in MODEL_WORLD.items():
            for s_key in m_val["scenarios"]:
                tasks.append((m_key, s_key))
        random.shuffle(tasks)

        bar = tqdm(tasks, desc=f"Ep {epoch+1}")

        for m_key, s_key in bar:
            gym = HinterlangGym(m_key, sim_class)

            # 1. Reset & Calculate Custom Baseline
            gym.reset(s_key)
            # We must calculate baseline using OUR new weights, not the gym's default
            base_df = gym.sim.run(silent_print=True)

            def get_hawk_loss(df):
                if "QPM" in m_key: c_col, g_col = "DLA_CPI", "L_GDP_GAP"
                else: c_col, g_col = "pinf", "y"

                # Look at 12 steps
                cpi = df[c_col].iloc[8:20].values
                gdp = df[g_col].iloc[8:20].values
                return np.sum((cpi**2)*W_CPI) + np.sum((gdp**2)*W_GDP)

            base_loss = get_hawk_loss(base_df)
            state = {k: round(float(base_df[k].iloc[8]), 2) for k in MODEL_WORLD[m_key]["targets"]}

            # 2. Prompt
            menu_pol = list(MODEL_WORLD[m_key]["mappings"].keys())
            prompt = (
                f"<|im_start|>system\n"
                f"You are a Central Bank AI. Crisis: {s_key}.\n"
                f"Goal: Minimize Inflation variance (Hawk Protocol).\n"
                f"Available Policies: {menu_pol}.\n"
                f"Valid Directions: [TIGHTEN, EASE, HOLD].\n"
                f"Valid Intensities: [MILD, MODERATE, AGGRESSIVE].\n"
                f"Output strictly in JSON.\n"
                f"<|im_end|>\n"
                f"<|im_start|>user\n"
                f"State: {json.dumps(state)}\n"
                f"Action JSON:\n<|im_end|>\n<|im_start|>assistant\n"
            )

            inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
            p_len = inputs.input_ids.shape[1]

            # 3. Generate
            model.eval()
            with torch.no_grad():
                outputs = model.generate(
                    **inputs, max_new_tokens=60, do_sample=True, temperature=1.4,
                    num_return_sequences=GROUP_SIZE, pad_token_id=tokenizer.eos_token_id
                )
            model.train()

            # 4. Score & Train
            rewards = []
            for i in range(GROUP_SIZE):
                text = tokenizer.decode(outputs[i][p_len:], skip_special_tokens=True)

                # --- FUZZY PARSER (Integrated) ---
                # (Paste your parser logic here or call the function)
                # Ensure it uses the NEW INTENSITY_SCALARS defined above!
                shocks = translate_semantic_action(text, m_key)

                if not shocks:
                    rewards.append(-2.0)
                    continue

                # --- STEP ---
                # Manually step to ensure we use Micro-Scalars
                sim_clone = gym.sim.clone()
                merged = copy.deepcopy(gym.current_scenario)
                for k, v in shocks.items():
                    if k not in merged: merged[k] = [0.0]*50
                    curr = v
                    for t in range(8, 12):
                        merged[k][t] += curr
                        curr *= 0.8

                sim_clone.reset_shocks()
                sim_clone.inject_shocks(merged)
                try:
                    df_act = sim_clone.run(silent_print=True)
                    act_loss = get_hawk_loss(df_act)

                    # --- FIX 3: ASYMMETRIC REWARD ---
                    delta = base_loss - act_loss
                    if delta > 0:
                        # Big reward for improvement
                        rew = np.tanh(delta / 10.0) * 5.0
                    else:
                        # Small penalty for failure (encourages trying)
                        rew = np.tanh(delta / 10.0) * 0.5

                    # --- OPTIONAL: HEURISTIC OVERRIDE ---
                    # If it's Stagflation and we TIGHTEN, give a small nudge
                    if "STAGFLATION" in s_key and "TIGHTEN" in text and rew < 0:
                        rew = 0.5 # "Good thinking, but physics failed you"

                    rewards.append(rew)
                except:
                    rewards.append(-5.0)

            # 5. Backprop
            rt = torch.tensor(rewards, device=model.device, dtype=torch.float32)
            epoch_rewards.extend(rewards)

            # Normalize advantages
            if torch.std(rt) < 1e-6: adv = torch.zeros_like(rt)
            else: adv = (rt - rt.mean()) / (rt.std() + 1e-8)

            optimizer.zero_grad()
            for i in range(GROUP_SIZE):
                seq = outputs[i].unsqueeze(0)
                out = model(seq)
                logits = out.logits[:, :-1, :]
                targets = seq[:, 1:]
                log_probs = F.log_softmax(logits, dim=-1).gather(2, targets.unsqueeze(-1)).squeeze(-1)
                mask = torch.zeros_like(log_probs)
                mask[:, p_len-1:] = 1.0
                loss = -(adv[i] * (log_probs * mask).sum()) / GROUP_SIZE
                loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
            optimizer.step()

            bar.set_description(f"Ep {epoch+1} | BestR: {max(rewards):.2f}")

    print(f"✅ Training Complete. Saved to {SAVE_DIR}")
    model.save_pretrained(f"{SAVE_DIR}/calibrated_model")

# EXECUTE
train_production_grpo_calibrated(model, tokenizer, GenericDSGESimulator)
````

````python
# ===================== CELL: CALIBRATION GRID SEARCH =====================
import numpy as np
import pandas as pd
import copy

# --- 1. DEFINE ROBUST CONFIG (Self-Contained) ---
CALIBRATION_WORLD = {
    "QPM_Full_Control": {
        "path": "model.yaml",
        "targets": ["DLA_CPI", "L_GDP_GAP"],
        "scenarios": {
            # Low Growth, Low Inf -> Needs CUT (-1.0)
            "RECESSION": {"SHK_L_GDP_GAP": [-3.0]*8, "SHK_DLA_CPI": [-1.0]*8},
            # High Inf, Low Growth -> Needs HIKE (+1.0)
            "STAGFLATION": {"SHK_DLA_CPI": [2.5]*8, "SHK_L_GDP_GAP": [-1.5]*8}
        }
    },
    "SW_Business_Cycles": {
        "path": "sw_model.yaml",
        "targets": ["pinf", "y"],
        "scenarios": {
            # High Inf -> Needs HIKE (+1.0)
            "COST_PUSH": {"epinf": [4.0]*8}
        }
    }
}

# --- 2. EXPERIMENT PARAMETERS ---

# A. ACTION MAGNITUDES (The "Action Space" Test)
# We test progressively larger shocks to see where the simulator breaks.
MAGNITUDE_GRID = [0.05, 0.10, 0.25, 0.50]

# B. LOSS WEIGHTS (The "Reward Function" Test)
WEIGHT_GRID = {
    "STANDARD": {"w_inf": 1.5, "w_gdp": 1.0}, # Current
    "HAWK":     {"w_inf": 3.0, "w_gdp": 0.1}, # Ignore GDP Pain
    "GROWTH":   {"w_inf": 0.5, "w_gdp": 2.0}, # Ignore Inflation Pain
}

def run_calibration_grid(sim_class):
    print(f"\n{'='*110}")
    print("🛠️  SIMULATION CALIBRATION GRID SEARCH")
    print("    Finding the 'Goldilocks Zone' for Action Size and Reward Weights")
    print(f"{'='*110}")

    # Headers
    print(f"{'SCENARIO':<20} | {'ACTION SIZE':<11} | {'WEIGHTING':<10} | {'BASE LOSS':<9} | {'ACT LOSS':<9} | {'DELTA':<8} | {'VALID?'}")
    print("-" * 110)

    # Scenarios to Test
    # Format: (Model, Scenario, Shock_Key, Direction_Mult, Description)
    TEST_CASES = [
        ("QPM_Full_Control", "RECESSION", "SHK_RS", -1.0, "Recession (Cut)"),
        ("QPM_Full_Control", "STAGFLATION", "SHK_RS", 1.0, "Stagflation (Hike)"),
        ("SW_Business_Cycles", "COST_PUSH", "em", 1.0, "Cost Push (Hike)")
    ]

    for model_key, scen_name, shock_var, direction, desc in TEST_CASES:

        # 1. Setup Simulation Once per Scenario
        try:
            conf = CALIBRATION_WORLD[model_key]
            # Path Logic
            p = conf["path"]
            import os
            if not os.path.exists(p) and os.path.exists(f"/content/{p}"): p = f"/content/{p}"

            sim = sim_class(p, periods=40, silence_snowdrop=True)
            scen_data = conf["scenarios"][scen_name]

            # Run Baseline
            base_sim = sim.clone()
            base_sim.reset_shocks()
            base_sim.inject_shocks(scen_data)
            df_base = base_sim.run(silent_print=True)

        except Exception as e:
            print(f"⚠️ SKIP {desc}: Sim Error ({str(e)})")
            continue

        # 2. Iterate Magnitudes
        for mag in MAGNITUDE_GRID:

            # Apply Action (Magnitude * Direction)
            act_sim = sim.clone()
            merged = copy.deepcopy(scen_data)

            if shock_var not in merged: merged[shock_var] = [0.0]*50

            # Inject with decay (Simulating the Agent's behavior)
            curr = mag * direction
            for t in range(8, 12):
                merged[shock_var][t] += curr
                curr *= 0.8

            act_sim.reset_shocks()
            act_sim.inject_shocks(merged)
            df_act = act_sim.run(silent_print=True)

            # 3. Iterate Weights
            if "QPM" in model_key: c_col, g_col = "DLA_CPI", "L_GDP_GAP"
            else: c_col, g_col = "pinf", "y"

            burn = 8
            horizon = 12

            for w_name, w_vals in WEIGHT_GRID.items():

                def get_loss(df):
                    c = df[c_col].iloc[burn:burn+horizon].values
                    g = df[g_col].iloc[burn:burn+horizon].values
                    # Simple Sum of Squares
                    return np.sum((c**2)*w_vals["w_inf"]) + np.sum((g**2)*w_vals["w_gdp"])

                l_base = get_loss(df_base)
                l_act = get_loss(df_act)
                delta = l_base - l_act

                # Validation Logic
                # If Delta > 0, the physics and rewards are aligned (Action worked)
                is_valid = "✅ PASS" if delta > 0 else "❌ FAIL"

                # Aesthetic filtering: Only print if it's a pass OR it's the standard failure
                # (To keep the table readable, we print everything for now)

                print(f"{desc:<20} | {mag:<11.2f} | {w_name:<10} | {l_base:<9.0f} | {l_act:<9.0f} | {delta:<8.1f} | {is_valid}")

        print("-" * 110)

run_calibration_grid(GenericDSGESimulator)
````

````python
# ===================== CELL 5: FINAL ROBUST TRAINING (ADAPTER PATTERN) =====================
import torch
import torch.nn.functional as F
from tqdm.auto import tqdm
import random
import numpy as np
import copy
import json

# --- 1. DEFINING THE PHYSICS ADAPTERS (Derived from your Audit) ---
# This ensures every model feels "fair" to the Agent.
PHYSICS_ADAPTERS = {
    "QPM_Full_Control": {
        "action_multiplier": 1.0,     # Bazooka level (Ratio 0.43)
        "reward_weight": 1.0,         # Standard Scale
        "mode": "HAWK"                # Focus on Inflation/Gap trade-off
    },
    "GSW_Pandemic_Sim": {
        "action_multiplier": 1.0,     # Bazooka level (Ratio 0.42)
        "reward_weight": 1.0,         # Standard Scale
        "mode": "CRISIS"              # Focus on Output Recovery
    },
    "SW_Business_Cycles": {
        "action_multiplier": 0.2,     # DOWN-SCALE (Fixing the 2.31 Ratio)
        "reward_weight": 5.0,         # Boost reward signal (SW values are small)
        "mode": "BALANCED"
    }
}

# --- 2. BASE ACTION SPACE (Normalized 0.0 to 5.0) ---
INTENSITY_SCALARS = {
    "MILD": 0.5, "LOW": 0.5, "SMALL": 0.5,
    "MODERATE": 1.5, "MEDIUM": 1.5, "STANDARD": 1.5,
    "AGGRESSIVE": 3.0, "HIGH": 3.0, "STRONG": 3.0,
    "EMERGENCY": 5.0, "EXTREME": 5.0, "MAX": 5.0
}

# --- 3. THE TRAINING LOOP ---
def train_final_adapter(model, tokenizer, sim_class):

    # Hyperparameters
    EPOCHS = 6 # 6 Epochs is enough given the clear signals
    GROUP_SIZE = 4
    LR = 5e-6
    MAX_GRAD_NORM = 1.0

    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)

    # Use the Calibrated World from your successful Audit
    TRAIN_WORLD = CALIBRATED_WORLD

    print(f"🚀 STARTING FINAL ADAPTIVE TRAINING")
    print(f"   MODELS: QPM (Macro), SW (BizCycle), GSW (Pandemic)")

    for epoch in range(EPOCHS):
        print(f"\n📅 === EPOCH {epoch+1} ===")

        # Create Tasks
        tasks = []
        for m_key, m_val in TRAIN_WORLD.items():
            for s_key in m_val["scenarios"]:
                tasks.append((m_key, s_key))
        random.shuffle(tasks)

        bar = tqdm(tasks, desc=f"Ep {epoch+1}")

        for m_key, s_key in bar:
            gym = HinterlangGym(m_key, sim_class)
            adapter = PHYSICS_ADAPTERS.get(m_key, {"action_multiplier": 1.0, "reward_weight": 1.0})

            # 1. BASELINE & STATE
            state, _ = gym.reset(s_key)
            base_df = gym.sim.run(silent_print=True)

            # --- ROBUST LOSS CALCULATOR ---
            def get_loss(df):
                # Auto-detect columns
                cols = df.columns
                # Inf/Price
                if "DLA_CPI" in cols: c_inf = df["DLA_CPI"]
                elif "pinf" in cols: c_inf = df["pinf"]
                else: c_inf = np.zeros(50)

                # Output/GDP
                if "L_GDP_GAP" in cols: c_out = df["L_GDP_GAP"]
                elif "y" in cols: c_out = df["y"]
                else: c_out = np.zeros(50)

                # Weights based on Adapter Mode
                w_inf, w_out = 1.0, 1.0
                if adapter["mode"] == "HAWK": w_inf, w_out = 3.0, 0.1 # Sacrifice GDP
                if adapter["mode"] == "CRISIS": w_inf, w_out = 0.5, 3.0 # Save GDP (Pandemic)

                # Terminal Loss (Steps 8-12)
                return (np.sum(c_inf[8:20]**2)*w_inf + np.sum(c_out[8:20]**2)*w_out)

            base_loss = get_loss(base_df)

            # 2. PROMPT
            menu = list(TRAIN_WORLD[m_key]["mappings"].keys())
            prompt = (
                f"<|im_start|>system\n"
                f"You are a Central Bank AI. Crisis: {s_key}.\n"
                f"Levers: {menu}. Directions: [TIGHTEN, EASE, HOLD]. Intensities: [MILD, AGGRESSIVE, EMERGENCY].\n"
                f"Objective: Minimize Economic Loss.\n"
                f"Output JSON.\n"
                f"<|im_end|>\n"
                f"<|im_start|>user\n"
                f"State: {json.dumps(state)}\n"
                f"Action JSON:\n<|im_end|>\n<|im_start|>assistant\n"
            )

            inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
            p_len = inputs.input_ids.shape[1]

            # 3. GENERATE
            model.eval()
            with torch.no_grad():
                outputs = model.generate(
                    **inputs, max_new_tokens=60, do_sample=True, temperature=1.3,
                    num_return_sequences=GROUP_SIZE, pad_token_id=tokenizer.eos_token_id
                )
            model.train()

            # 4. STEP
            rewards = []
            for i in range(GROUP_SIZE):
                text = tokenizer.decode(outputs[i][p_len:], skip_special_tokens=True)

                # Parse
                raw_shocks = translate_semantic_action(text, m_key)

                if not raw_shocks:
                    rewards.append(-5.0)
                    continue

                # --- APPLY ADAPTER MULTIPLIER ---
                # This fixes the "OP" vs "Weak" problem
                mult = adapter["action_multiplier"]
                final_shocks = {k: v * mult for k, v in raw_shocks.items()}

                # Simulate
                sim_clone = gym.sim.clone()
                merged = copy.deepcopy(gym.current_scenario)
                for k, v in final_shocks.items():
                    if k not in merged: merged[k] = [0.0]*50
                    curr = v
                    for t in range(8, 12):
                        merged[k][t] += curr
                        curr *= 0.8

                sim_clone.reset_shocks()
                sim_clone.inject_shocks(merged)

                try:
                    df_act = sim_clone.run(silent_print=True)
                    act_loss = get_loss(df_act)

                    # Reward Logic
                    delta = base_loss - act_loss

                    # Apply Reward Weight (Boosts SW signals)
                    weighted_delta = delta * adapter["reward_weight"]

                    # Tanh Clipping (Prevents gradients from exploding)
                    rew = np.tanh(weighted_delta / 5.0) * 10.0

                    # Bonus for merely trying in a Crisis (Overcoming "Freeze")
                    if "CRISIS" in adapter["mode"] and rew > -2.0 and "HOLD" not in text:
                        rew += 1.0

                    rewards.append(rew)
                except:
                    rewards.append(-5.0)

            # 5. BACKPROP
            rt = torch.tensor(rewards, device=model.device, dtype=torch.float32)
            if torch.std(rt) < 1e-6: adv = torch.zeros_like(rt)
            else: adv = (rt - rt.mean()) / (rt.std() + 1e-8)

            optimizer.zero_grad()
            for i in range(GROUP_SIZE):
                seq = outputs[i].unsqueeze(0)
                out = model(seq)
                logits = out.logits[:, :-1, :]
                targets = seq[:, 1:]
                log_probs = F.log_softmax(logits, dim=-1).gather(2, targets.unsqueeze(-1)).squeeze(-1)
                mask = torch.zeros_like(log_probs); mask[:, p_len-1:] = 1.0
                loss = -(adv[i] * (log_probs * mask).sum()) / GROUP_SIZE
                loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
            optimizer.step()

            bar.set_description(f"Ep {epoch+1} | {m_key[:4]} | R: {max(rewards):.2f}")

    print(f"✅ FINAL TRAINING COMPLETE.")
    model.save_pretrained(f"{SAVE_DIR}/production_model_v1")
    tokenizer.save_pretrained(f"{SAVE_DIR}/production_model_v1")

# EXECUTE
train_final_adapter(model, tokenizer, GenericDSGESimulator)
````
